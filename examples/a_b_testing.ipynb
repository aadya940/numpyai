```python
from dotenv import load_dotenv
load_dotenv(".env")

import numpyai as npi
import numpy as np
```

```python
np.random.seed(42)  # For consistent results
```

```python
# Simulate conversions for group A (10% conversion rate)
a_data = np.random.binomial(n=1, p=0.10, size=1000)

# Simulate conversions for group B (12% conversion rate)
b_data = np.random.binomial(n=1, p=0.12, size=1000)
```

```python
# Start a NumPyAI session with both arrays
sess = npi.NumpyAISession([a_data, b_data], verbose=True)
```

```python
print(sess.chat("What is the conversion rate of each group?"))  # → Should return two means
```

```python
# Chat: Perform a z-test to compare the conversion rates
print(sess.chat("Perform a z-test to determine if the difference is statistically significant."))
```

```python
print(sess.chat("Which group performed better, and is the difference significant?"))
```
